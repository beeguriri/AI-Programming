```python
str1 = '  Python is wonderful!  '
print(str1.upper())
print(str1.lower())
print(str1.title())
print(str1.lstrip())
```

```python
# String의 min/max : string내 char의 사전순서로 비교
# ex) min : string내 제일 빠른 글자
# 대문자가 우선순위 높음
str1 = 'pythonisgood'
str2 = 'Pythonisxllnt'
print(min(str1))
print(max(str1))
print(min(str2))
print(max(str2))
print(min(str1,str2))
print(max(str1,str2))
```

```python
str1 = 'python is good'
print(str1[:5])
print(str1[5:])
```

```python
print(str1.find('i'))
print(str1[str1.find('i')])
```

```python
str1[::-1]
```
