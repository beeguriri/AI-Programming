```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3],[1,4,9])
plt.plot([1,2,3,4],[1,6,7,8])
plt.xlim(1,4)
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.legend(['a', 'b'])
plt.show()
```

```python
fileName = '../data/firstChoice.txt'
fileName2 = '../data/SA.txt'

def read_data(fileName):
  result = list()
  with open(fileName, 'r') as f_read:
    for line in f_read:
      result.append(float(line))
  return result

# Search performance by First-choice hill climbing for TSP
x = np.arange(len(read_data(fileName))) # There are 14140 values to be plotted
x2 = np.arange(len(read_data(fileName2)))
y = read_data(fileName) # Y values will be read in from a file
y2 = read_data(fileName2)

plt.plot(x, y)
plt.plot(x2, y2)
plt.xlabel('Number of Evaluations')
plt.ylabel('Tour Cost')
plt.title('TSP-100 by First-Choice Hill Climbing')
plt.show()
```

```python
# Plot for a Gaussian function
def gaussian(x):
  return np.exp(-(0.5 - x)**2/1.5)

x = np.linspace(-2, 2.5, 100)    # x = np.arange(-2, 2.5, 0.01)
y = gaussian(x) # Array argument produces array output
plt.plot(x, y)
plt.xlabel('x values')
plt.ylabel('exp(-(0.5-x)**2/1.5)')
plt.title('Gaussian Function')
plt.show()
```

## Figure

```python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
plt.show()
```

```python
fig = plt.figure()  # an empty figure with no Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
plt.show()
```

```python
fig = plt.figure()  # an empty figure with no Axes
# a figure with one axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right-top'],
                               ['left', 'right_bottom']])
plt.show()
```

## Axis

```python
x = [1, 2, 3, 4]

y1 = [1, 4, 2, 3]
y2 = [2, 3, 4, 6]

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html
fig, ax = plt.subplots()  # Create a figure containing a single axes.

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
ax.plot(x, y1, label='line 1')  # Plot some data on the axes.
ax.plot(x, y2, label='line 2')  # Plot some data on the axes.

ax.set_xticks(np.arange(1, 5, 1))
ax.set_xlabel('X axis label')
ax.set_ylabel('Y axis label')
ax.set_title('two plots')

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
ax.legend(loc='upper left')

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html
ax.grid(linestyle='--')
plt.show() # the current figure is displayed and then emptied
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
xdata = np.arange(len(data1))  # make an ordinal for this
data = 10**data1

fig, axs = plt.subplots(1, 2, figsize=(8, 3), layout='constrained')
axs[0].plot(xdata, data)

# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylim.html
axs[0].set_ylim(0, 10**3)

# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_yscale.html
axs[1].set_yscale('log')
axs[1].plot(xdata, data)
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, layout='constrained')
axs[0].plot(xdata, data1)
axs[0].set_title('Automatic ticks')

axs[1].plot(xdata, data1)
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xticks.html
axs[1].set_xticks(ticks=np.arange(0, 100, 30), labels=['zero', '30', 'sixty', '90'])
axs[1].set_yticks([-1.5, 0, 1.5])  # note that we don't need to specify labels
axs[1].set_title('Manual ticks')
plt.show()
```
